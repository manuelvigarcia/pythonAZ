# Automatización

```python
raw_data['Area'].unique()
```

```python
contrataciones_avg = []
for area in sorted(raw_data['Area'].unique()):
    for grupo in sorted(raw_data[raw_data['Area']==area]['Grupo'].unique()):
        for nivel in sorted(raw_data[(raw_data['Area']==area) & (raw_data['Grupo']==grupo)]['Nivel'].unique()):
            area_grupo_nivel_sueldos=raw_data[(raw_data['Area']==area) & (raw_data['Grupo']==grupo) & (raw_data['Nivel']==nivel)]['Retribucion']
            print(f'{area},{grupo},{nivel}: {area_grupo_nivel_sueldos.count()} -> {round(area_grupo_nivel_sueldos.mean())}')
            fila = [area,grupo,nivel,area_grupo_nivel_sueldos.count(),round(area_grupo_nivel_sueldos.mean())]
            contrataciones_avg.append(fila)
```

```python
contrataciones_avg
```

```python
calor = pd.DataFrame(contrataciones_avg)
```

```python
calor.columns.values
```

```python
columns = ['Area','Grupo','Nivel','Contrataciones','Salario Medio']
calor.columns = columns
```
