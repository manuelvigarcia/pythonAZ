```python
import numpy as np
from numpy.random import randn
```

```python
randn()
```

```python
# if
x = randn()
answer = None
if x > 1:
    answer= "greater than 1"
print(x)
print(answer)
```

```python
# else
x = randn()
answer = None
if x > 1:
    answer= "greater than 1"
else:
    answer = "Less than 1"
print(x)
print(answer)
```

```python
# nested if
x = randn()
answer = None
if x > 1:
    answer= "greater than 1"
else:
    if x>=-1:
        answer = "Between -1 and 1"
    else:
        answer = "Less than -1"
print(x)
print(answer)
```

```python
# chained if
x = randn()
answer = None
if x > 1:
    answer= "greater than 1"
elif x >= -1:
    answer = "Between -1 and 1"
else:
    answer = "Less than -1"
print(x)
print(answer)
```
