# The "While" Loop

Python is not vectorised as R, it is a object oriented language

```python
# how it looks in any other language
# while(condition) {
#   statement1
#   statement2
# }

# In Python (indentation is important, so you don't need the curly braces
# while condition:
#    statement1
#    statement1

while False:
    print ("Hello")
```

```python
counter = 0
while counter < 12:
    print(counter)
    counter = counter + 1
print("Hello")
```

```python
# Exercise: create the for outputs with a while loop
counter = 0
while counter < 5:
    print(f"hello python ({counter})")
    counter = counter + 1
    
```

```python
counter = 0
while counter < 5:
    print("hello python: ", counter)
    print("hello line")
    counter = counter + 1
```

```python
mylist = [10, 100, 1000]
jj = 0
while jj < len(mylist):
    print ("jj is now: ", mylist[jj])
    jj = jj + 1
```
