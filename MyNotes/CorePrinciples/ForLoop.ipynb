# The For Loop

```python
range(5)
```

```python
list(range(0,5))
```

```python
for i in range(5):
    print(f"hello python ({i})")
```

---

```python
for i in range(5):
    print("Hello python: ", i)
    print("Hello line")
```

```python
mylist = [10,100,1000]
for jj in mylist:
    print("jj now is: ", jj)
```

The for loop can also be used to perform actions in a vector

```python
strings = ['word','concept','symbol','idea','quality']
my_strings = [('my_' + i) for i in strings]
my_strings
```

This can be used to operate the vector or matrix in whichever way needed.
