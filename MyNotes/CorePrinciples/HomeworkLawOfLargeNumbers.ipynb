# Law of large numbers
The average of the sample converges to the expected value of the measured variable

## Assignment
Test the Law fo Large Numbers for N random normally distributed numbers with mean 0 and stdev 1:
- create a python script that will count how many of these numbers fall between -1 and 1 and divide by the total quantity N of numbers drawn
- According to the normal distribution, the result shall be 68.2%
- Check that the mean converges to the expected value as N increases.

```python
import numpy as np
from numpy.random import randn
```

```python
N = 1
while N < 100000000:
    print("N is ", N)
    normales = 0
    for i in range(N):
        x = randn()
        if (x > -1 and x < 1):
            normales = normales + 1
    print(f"From {N} drawn numbers, {normales} fell between -1 and 1; that is {normales/N*100}%")
    N = N * 10
```
