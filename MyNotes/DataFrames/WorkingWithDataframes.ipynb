# Filtering Data Frames
You filter rows by column values

```python
# Evaluate which rows have (internetUsers<2) == true
stats.InternetUsers<2
```

```python
 # Create a Filter
Filter = stats.InternetUsers<2  # --> this is a list of True|False values that can be used for slicing rows of the DataFrame
```

```python
stats[Filter]
```

```python
# Countries with more than 40 births per 1000 inhabitants per year
Filter2 = stats.BirthRate>40
stats[Filter2]
```

```python
#combine two filters
stats[Filter & Filter2]
```

# Accessing individual elements / cells

```python
# You are rarely going to need one single specific value from a Data Frame. But here it goes
# .at  -> for labels
# .iat -> for integer location
stats.iat[3,4]
```

```python
stats.head()
```

```python
stats.iat[0,1]
```

```python
stats.at[2,'BirthRate'] # Keep in mind that 2 happens to be the label for the third row.
```

```python
# How are .iat and at different?
```

```python
sub10 = stats[::10]
sub10
```

```python
sub10.iat[10,0] # gives the 11th row: libya
```

```python
sub10.at[10,'CountryName'] # works with labels, i.e. the row with label 10 as row index: Azerbaijan
```

The confusion comes because, unless you alter row order, or take out rows, the "index" label tends to coincide with the row ordinal in the DataFrame

# Visualization: Seaborn

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
plt.rcParams['figure.figsize']=12,4
```

```python
vis1 = sns.histplot(stats['InternetUsers'], bins=20)
plt.show()
```

```python
vis2=sns.boxplot(data=stats, x='IncomeGroup', y='BirthRate')
plt.show()
```

### Plot internet users vs. birth rate.

```python
vis3=sns.lmplot(x='InternetUsers', y= 'BirthRate', data=stats, fit_reg=False)
plt.show()
```

```python
vis4=sns.lmplot(x='InternetUsers', y= 'BirthRate', data=stats, fit_reg=False, hue='IncomeGroup')
plt.show()
```

```python
#marker size and alpha are parameters of pyplot, not lmplot, so you can pass them through with the 'keyword' parameters feature
vis4=sns.lmplot(x='InternetUsers', y= 'BirthRate', data=stats, fit_reg=False, hue='IncomeGroup', scatter_kws={'s':100,'alpha':0.4})
plt.show()
```
