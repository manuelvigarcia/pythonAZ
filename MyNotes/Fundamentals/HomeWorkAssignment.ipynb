```python
import numpy as np
revenue = [14574.49, 7606.46, 8611.41, 9175.41, 8058.65, 8105.44, 11496.28, 9766.09, 10305.32, 14379.96, 10713.97, 15433.50]
expenses = [12051.82, 5695.07, 12319.20, 12089.72, 8658.57, 840.20, 3285.73, 5821.12, 6976.93, 16618.61, 10054.37, 3803.96]
```

```python
#profit for each month
profit = np.subtract(revenue,expenses)
profit
```

```python
# profit after tax of 30%
profit_after_tax = profit * 0.7
profit_after_tax
```

```python
# profit margin for each month - equals to profit after tax divided by revenue
profit_margin = np.divide(profit_after_tax, revenue)*100
profit_margin
```

```python
# good months - where the profit after tax was greater than the mean for the year
good_months = profit_after_tax > profit_after_tax.mean()
good_months
```

```python
# bad months - where the profit after tax was less than the mean for the year
# I do not agree with these being the oposite to good_months: one month can be an average month
bad_months = profit_after_tax < profit_after_tax.mean()
bad_months
```

```python
# the best month - where the profit after tax was max for the year
best_month = profit_after_tax == profit_after_tax.max()
best_month
```

```python
# the worst month - where the profit after tax was min for the year
worst_month = profit_after_tax == profit_after_tax.min()
worst_month
```
