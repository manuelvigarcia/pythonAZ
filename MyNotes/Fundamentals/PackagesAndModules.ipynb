# Packages and Modules

1. Find which package you need: search in github 
2. install the package
   
     a) open the anaconda command prompt
   
     b) conda install packagename
   
        or (if you have anaconda environment)
   
        pip install packagename
   
4. import packagename
   
   or
   
   from packagename import module
   
   or
   
   from module import function
