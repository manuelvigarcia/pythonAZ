## Slicing
The operator ':' has a "from" before it and an "to" after it.

```python
mylist = ['A','B','C','D','E','F','G','H','I','J']
```

```python
mylist[3]
```

```python
mylist[7]
```

```python
mylist[3:7]
```

It ends before the last element indicated.

```python
mylist[3:7:2]
```

. . . whith a step 2, the last argument
