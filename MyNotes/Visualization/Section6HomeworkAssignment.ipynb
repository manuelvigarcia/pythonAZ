```python
from matplotlib import pyplot as plt
import seaborn as sns
sns.set_style("darkgrid")
%matplotlib inline
```

Identify the 5 genres with most movies to build a plot of them

```python
dataset.Genre.value_counts().head(5)
```

```python
genres=['action','comedy','adventure','animation','drama']
```

```python
studios=['Buena Vista Studios', 'Sony', 'Universal',
       'Paramount Pictures', 'WB', 'Fox']
```

```python
alldata = dataset[(dataset.Studio.isin(studios)) & (dataset.Genre.isin(genres))]
```

```python
alldata.Genre.unique()
```

```python
len(genres)
```

```python
fig1=sns.boxplot(data=alldata,x='Genre',y='USGross%', color='lightgray', fliersize=0.0)
sns.stripplot(data=alldata,x='Genre', y='USGross%',hue='Studio',alpha=0.4)
plt.legend(loc='upper left', bbox_to_anchor=(1,1), frameon=False)
plt.title("Domestic Gross % by Genre", fontsize=25)
plt.ylabel("Gross % US", fontsize=15)
plt.xlabel("Genre", fontsize=15)
plt.show()
```
