```python
# Ad data labels
list1 = list()
labels = list()
for gen in movies.Genre.cat.categories:
    list1.append(movies[movies.Genre == gen].BudgetMillions)
    labels.append(gen)

h = plt.hist(list1, bins=30,stacked=True,rwidth=1,label=labels)
plt.legend()
plt.show()
```

## Show the water of planet earth versus the planet itself

```python
#volumen = 4/3 * pi *r^3
from math import pi
def volumen_esfera(radio=1):
    return 4/3*pi*(radio**3)
def area_esfera(radio=1):
    return 4*pi*(radio**2)
def area_circulo(radio=1):
    return pi*(radio**2)
```

```python
volumen_esfera()
```

```python
earth_volume_km3=1083210000000
earth_diameter_km=6378.1*2
```

```python
water_volume_km3=1386000000
water_diameter_km=1386
```

```python
volumen_esfera(earth_diameter_km/2)
```

```python
area_oceanos=area_esfera(earth_diameter_km/2)*0.71
area_oceanos
```

```python
profundidad_oceanos = water_volume_km3/area_oceanos
profundidad_oceanos
```

3,819 meters is the average depth of the ocean if all water were in the 71% of the surface of the earth.

```python
earth_volume_km3/volumen_esfera(earth_diameter_km/2)
```

```python
water_volume_km3/volumen_esfera(water_diameter_km/2)
```

```python
# difficult to paint such big dimensions. Normalize
earth_norm_radius = 1
water_norm_radius = water_diameter_km/earth_diameter_km

fig, ax = plt.subplots()
earth_circle=plt.Circle((0,0),earth_norm_radius,fill=False)
water_circle=plt.Circle((0.79,0.79),water_norm_radius,fill=False)
ax.add_patch(earth_circle)
ax.add_patch(water_circle)
ax.set_aspect('equal')
ax.set_xlim(-1,1)
ax.set_ylim(-1,1)
plt.show()
```

## Work with facet grid
One graphs distributed in several visualizations, so the data are not that crammed.

```python
plt.scatter(movies.CriticRating,movies.AudienceRating)
plt.show()
```

```python
g = sns.FacetGrid(movies,row='Genre', col='Year', hue='Genre')
g = g.map(plt.scatter, 'CriticRating', 'AudienceRating')
plt.show()
```

```python
g = sns.FacetGrid(movies,row='Genre', col='Year', hue='Genre')
g = g.map(plt.hist, 'BudgetMillions')
plt.show()
```

```python
g = sns.FacetGrid(movies,row='Genre', col='Year', hue='Genre')
kws={'s':50,'linewidth':0.5,'edgecolor':'black'}
g = g.map(plt.scatter, 'CriticRating', 'AudienceRating',**kws)
plt.show()
```

### adding diagonals and limiting axis

```python
g = sns.FacetGrid(movies,row='Genre', col='Year', hue='Genre')
kws={'s':50,'linewidth':0.5,'edgecolor':'black'}
g = g.map(plt.scatter, 'CriticRating', 'AudienceRating',**kws)
g.set(xlim=(0,100))
for ax in g.axes.flat:
    ax.plot((0,100),(0,100), c='gray',ls='--')
g.add_legend()
plt.show()
```
