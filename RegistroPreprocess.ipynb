```python
import pandas as pd
```

```python
dirpath='C:/Users/mvgarcia/Downloads/ComiteEmpresa/cb/2024/'
raw_data = pd.read_csv(dirpath + 'Registro_raw.csv')
raw_data
```

```python
def limpia_sueldo(x):
    return x.replace('.', '').replace(',', '.')
```

```python
raw_data.info()
```

```python
data = raw_data.copy()
```

```python
data['Id']=[(data['Fecha'][i]+f'{i:03d}') for i in range(0, len(data))] 
```

```python
data['Retribucion']=raw_data['Retribucion'].apply(limpia_sueldo).astype('float')
```

```python
data
```

```python
data['Area']=[i.split()[1] for i in data['Categoria']]
data['Grupo']=[i.split()[3] for i in data['Categoria']]
data['Nivel']=[i.split()[5] for i in data['Categoria']]
```

```python
preprocessed=data[['Id', 'Fecha', 'Retribucion', 'Area', 'Grupo', 'Nivel']]
```

```python
preprocessed.to_csv(dirpath + 'Preprocessed.csv', index = False)
```
